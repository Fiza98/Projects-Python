**B) Classification Using Naïve Bayes for: “Age”, “Nursery”, “Dalc”, “Walc” to
predict whether student wants to go to higher education found in the attribute:
“higher”**

```python
#1) Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#2) Import dataset
ds=pd.read_csv ('student-mat v2.csv')
ds.head()
```

```python
ds.info()
```

```python
#3) Allocate the relevant attributes as input and output
ds=ds.loc[:,['age','nursery','Dalc','Walc','higher']]
x=ds.iloc[:,:-1].values
y=ds.iloc[:,-1].values
```

```python
#4) Use LabelEncoder to encode categorical data
i=1
for i in range(1,2):
    d=list(set(ds.iloc[:, i].values))
    d.sort()
    print(i,d)
    i=i+1
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder_x= LabelEncoder()
x[:, 1] =labelencoder_x.fit_transform(x[:,1])
print(x)
```

```python
#5) Split your data into training and test sets with the appropriate proportions
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split (x, y, test_size = 0.25, random_state =0)
x_train , x_test , y_train , y_test
```

```python
#6) Normalized your data using StandardScaler
from sklearn.preprocessing import StandardScaler
sc_x=StandardScaler()
x_train=sc_x.fit_transform(x_train)
x_test=sc_x.transform(x_test)
```

```python
#7) Fit the and predict results using the Naïve Bayes Classifier
from sklearn.naive_bayes import BernoulliNB
classifier=BernoulliNB()
classifier=classifier.fit(x_train,y_train)
```

```python
#Predict the test set results
y_pred=classifier.predict(x_test)
y_test, y_pred
```

```python
#8) Evaluate your results using confusion matrix and calculate the prediction accuracy
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
#Finding accuracy
from sklearn.metrics import accuracy_score
AR=accuracy_score(y_test,y_pred)
print(AR)
```

```python
prediction=classifier.predict ([[20,1,3,2]])#age = 20, nursery = 1, dalc = 3, walc = 2
print(prediction)
```

**If the student aged 20, attended nursery school, workday alcohol consumption = 3, weekend alcohol consumption = 2, then the student wants to take higher education**

second(G2) grade**

```python
#1) Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#2) Import dataset
ml=pd.read_csv('student-mat v2.csv')
ml.head()
```

```python
ml.info()
```

```python
#3) Allocate the relevant attributes as input and output
ml=ml.loc[:,['G1','G2','G3']]
x1=ml.iloc[:,:-1].values 
y1=ml.iloc [:,-1].values
ml
```

```python
print(ml.corr())
ml.shape
```

```python
#4) Split your data into training and test sets with the appropriate proportions
from sklearn.model_selection import train_test_split
x1_train, x1_test , y1_train , y1_test = train_test_split (x1, y1, test_size = 0.25, random_state = 0)
x1_train, x1_test , y1_train , y1_test
```

```python
#import scaler
from sklearn.preprocessing import StandardScaler, MinMaxScaler
sc_X1 = StandardScaler()
x1_train = sc_X1.fit_transform(x1_train)
```

```python
#5) Fit the and predict results using the Multiple Linear Regression
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x1_train,y1_train)
print('Coefficients,beta:', regressor.coef_)
print('Constant,alpha:', regressor.predict([[0,0]]))
```
