**APRIORI - SUPPORT, CONFIDENCE, LIFT**

•Dataset – 7500 grocery transaction record

•Store located in an European country

•Motivation – look for associations between items

**the higher the lift, the higher association between them.**

```python
#apriori
#importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#data preprocessing
df = pd.read_csv('Market_Basket_Optimisation.csv', header = None)
df.head()
```

```python
#Replacing NaN with zeros
df.fillna(0, inplace = True)
df.head()
```

```python
#transformation
#0 to 19, but in programming add on 1, so 0 to 20

transactions = []
for i in range(0, len(df)):
    transactions.append([str(df.values[i,j]) for j in range(0, 20) if str(df.values[i, j])!='0'])
```

```python
transactions[0]

#each row takes every cell and translated into string, the put into list, then append for transaction 1 until end.
#if the string is 0, it will be NaN > refer to coding above
```

```python
#the min_support = 0.003, min_confidence = 0.2, min_lift = 3, max_length = 3 just try an error for this data
#training Apriori on the dataset
from apyori import apriori
rules = apriori(transactions, min_support = 0.003, min_confidence = 0.2, min_lift = 3, max_length = 3)# if u use max_length 2, u will not get the pairing for threee, only two

#saving the rules
results = list(rules)
```

```python
#save transactions and export the results as csv
df1 = pd.DataFrame(transactions)
df1.to_csv("transaction.csv")
df2 = pd.DataFrame(results)
df2.to_csv("results.csv")

results = pd.read_csv('results.csv')
results.head(10)
```
